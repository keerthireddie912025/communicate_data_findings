# Dataset Overview and Executive Summary
The Titanic dataset contains information for 891 passengers and 15 variables.

The dataset includes demographic information, passenger class, fare, embarkation details, and survival status.

This presentation highlights the most important findings discovered during exploratory data analysis using a small number of polished visualizations.

The objective of this presentation is to identify the passenger characteristics that most influenced survival during the Titanic disaster and communicate the findings using clear explanatory visualizations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style("whitegrid")
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv("dataset.csv")

df.head()
```

# Visualization 1

### Survival Rate by Gender

This visualization compares the survival rate of male and female passengers. It highlights the significant impact of gender on survival during the Titanic disaster.

```python
plt.figure(figsize=(7,5))

sns.barplot(
    data=df,
    x="sex",
    y="survived",
    estimator=np.mean,
    errorbar=None,
    palette=["steelblue","lightcoral"]
)

plt.title("Survival Rate by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Survival Rate")

for i, value in enumerate(df.groupby("sex")["survived"].mean()):
    plt.text(i, value + 0.02, f"{value:.2f}", ha="center", fontsize=11)

plt.ylim(0,1)

plt.show()
```

### Observation

Female passengers had a much higher survival rate than male passengers. This suggests that gender was one of the strongest factors influencing survival during the Titanic disaster.

# Visualization 2
### Fare Distribution by Passenger Class
This visualization compares the fare paid by passengers in each travel class.

# Visualization 2
### Fare Distribution by Passenger Class
This visualization compares the fare paid by passengers in each travel class.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="class",
    y="fare"
)

plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.show()
```

### Observation

First Class passengers generally paid much higher fares than passengers in Second and Third Class. The wider fare distribution in First Class reflects greater ticket price variation among wealthier passengers.

# Visualization 3
### Age vs Fare by Survival Status
This visualization explores how age and fare relate to passenger survival.

```python
plt.figure(figsize=(9,6))

sns.scatterplot(
    data=df,
    x="age",
    y="fare",
    hue="survived",
    palette=["red","green"],
    alpha=0.6
)

plt.title("Relationship between Age, Fare and Survival", fontsize=14)

plt.xlabel("Age (Years)")
plt.ylabel("Fare Paid")

plt.xlim(0,80)
plt.ylim(0,300)

plt.legend(
    title="Survival Status",
    labels=["Did Not Survive","Survived"]
)

plt.annotate(
    "Higher fares\nmostly survived",
    xy=(38,180),
    xytext=(45,240),
    arrowprops=dict(arrowstyle="->")
)

plt.figtext(
    0.02,
    -0.04,
    "Note: Fare values above 300 exist in the dataset but are hidden for readability.",
    fontsize=10
)

plt.show()
```
